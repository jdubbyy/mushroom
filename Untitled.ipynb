```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for piecewise horizontal lines
x_coords = [0, 2, 4, 6, 8, 10]
y_coords = [1, 3, 2, 4, 3, 1]

# Define the data for vertical lines
vertical_lines = [1, 3, 5, 7, 9]

# Define the data for shaded regions
shaded_regions = [(1, 2), (5, 6), (8, 9)]

# Create the plot
fig, ax = plt.subplots()

# Plot piecewise horizontal lines
for i in range(len(x_coords) - 1):
    ax.hlines(y_coords[i], x_coords[i], x_coords[i+1], colors='blue', linewidth=2)

# Plot vertical lines
for x in vertical_lines:
    ax.axvline(x, color='red', linestyle='--', linewidth=1)

# Plot shaded regions
for start, end in shaded_regions:
    ax.axvspan(start, end, facecolor='gray', alpha=0.3)

# Set plot labels and title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_title("Piecewise Horizontal Lines with Vertical Lines and Shaded Regions")

# Show the plot
plt.show()
```
